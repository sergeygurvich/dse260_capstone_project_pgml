# Explore the Ensemble JSON File

```python
with open('uniform-pgml-success.bp3d.json') as f:
    s = f.read()
ens_json = json.loads(s)
ens_json.keys()
```

```python
ens_json['type']
```

```python
# load into Pandas
ens_df_json = json.loads(ens_json['df'])
ens_df = pd.DataFrame().from_dict(ens_df_json)
ens_df.head(5)
```

## EXPLORE INPUTS: fuel, wind_speed, wind_direction, surface_moisture, topo, ignition

- Topography (1 unique value)
